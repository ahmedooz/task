# Grouping: Quantity of Each product per Customer

```python
s_test = s_df[['CUSTOMERNAME','PRODUCTLINE','QUANTITYORDERED']]
s_df_group = s_test.groupby(['PRODUCTLINE','CUSTOMERNAME'],as_index =False).mean()
```

```python
s_df_group
```

# Pivot table (Grouping: Quantity of Each product per Customer)

```python
s_df_pivot = s_df_group.pivot(index='CUSTOMERNAME',columns ='PRODUCTLINE')
```

```python
s_df_pivot
```

# ANOVA 

```python
df_anova=s_df[['PRODUCTLINE','PRICEEACH']]
grouped_anova = df_anova.groupby(['PRODUCTLINE'])
anova_result1=stats.f_oneway(grouped_anova.get_group('Motorcycles')['PRICEEACH'],grouped_anova.get_group('Ships')['PRICEEACH'])
```

```python
anova_result1
#between Motorcycle price Vs. Ships Price
```

The prices between Motorcycles and Ships are NOT significantly different.
